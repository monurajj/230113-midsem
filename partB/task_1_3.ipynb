# Task 1.3: What the Paper Claims to Improve

- **Baseline:** The main baseline or prior method compared against is the Discriminatively Trained Part Based Models by Felzenszwalb et al. (referred to in the paper as the "UoCTTI" 2-layer model).

- **Identified Limitation:** The paper identifies that the Felzenszwalb baseline uses a shallow (2-layer) structure, missing out on richer shape and appearance descriptions. Furthermore, expanding the baseline's approach to deeper structures is blocked by its complicated 3-stage layer-wise training procedure that requires careful, manual part selection and elaborate model initialization, which scales poorly in computational complexity.

- **Proposed Improvement:** The paper proposes replacing the complex, multi-stage heuristic training with an elegant, simultaneous multi-layer learning algorithm based on an Incremental Concave-Convex Procedure (iCCCP). This avoids the need for careful initialization (using a simple, generic grid topology) while massively speeding up the negative data mining process, making it feasible to train deeper 3-layer models.

- **Failure Condition:** The proposed method would **not** outperform the UoCTTI baseline on extremely low-resolution imagery or objects largely defined by a single macro-texture (e.g., a blurry ball or a simple cloud). In such conditions, adding a 3rd layer of 6x6 high-resolution internal parts would essentially fit to noise, adding 10,000+ unnecessary parameters. The data scarcity or lack of high-frequency detail would lead the deeper iCCCP model to overfit, whereas the simpler 2-layer Felzenszwalb model would be more robust.
