# Task 1.2: Key Assumptions

## Assumption 1
- **Assumption:** The spatial deformation relationship between object parts can be strictly modeled as a tree structure with locally independent parent-child offsets.
- **Why the method needs it:** The exact and efficient inference of the optimal part locations depends on dynamic programming. The recursive formulation structurally prohibits loopy dependencies (where sibling parts directly constrain each other) to ensure exact inference is tractable.
- **Violation scenario:** An object undergoing complex articulation or occlusion where two sibling parts physically interact or occlude each other (e.g., a person crossing their arms, where the left hand and right hand have strong spatial dependency that the parent "torso" node cannot mediate).
- **Paper reference:** Section 3.2, Equation (5) (the recursive dynamic programming equation summing over children nodes independent of other branches).

## Assumption 2
- **Assumption:** Deeper structural models (3 layers) with fixed, standard top-down topologies (1x1, 3x3, 6x6 grids) are universally beneficial and generic enough to capture object variances without class-specific custom topology design.
- **Why the method needs it:** The method avoids the complicated feature selection, careful part selection, and manual initialization steps characteristic of prior work. It simply instantiates the same fixed deep hierarchical tree for every class and relies on the SVM to zero out useless weights.
- **Violation scenario:** Highly non-rigid or heavily self-mutilating objects (like a snake or a rope) where a rigid 3x3/6x6 decomposition is completely inappropriate and semantic meaning cannot be aligned to a standardized rigid bounding-box grid.
- **Paper reference:** Section 3.1 ("The numbers of layers and nodes are the same for different object classes and views") and Section 4.3 ("different structures do not affect the performance when iCCCP has converged").

## Assumption 3
- **Assumption:** Optimal separating hyperplanes in the early stages of learning are directionally similar to the true optimum, meaning hard negative examples found in a small subset of background images are sufficient to constrain the initial weight updates.
- **Why the method needs it:** The core contribution of the iCCCP algorithm is expanding the negative set incrementally. If the early iterations using a tiny fraction of negative images updated the weights in a completely orthogonal or destructive direction, the iterative growth strategy would fail to converge to a global structural pattern.
- **Violation scenario:** A dataset where simple, frequent backgrounds totally mimic the target object's texture, confusing the SVM early on if a representative diversity of "hardest" negatives is not present in the small starting pool, leading the model to a poorly initialized local minimum that CCCP cannot escape.
- **Paper reference:** Section 3.4 (Incremental CCCP for Large Scale Training) & Figure 4 (iCCCP algorithm adding images iteratively).
